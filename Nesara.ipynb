```python
import importlib
libraries=['cv2','numpy','matplotlib']
def check_library(cv2):
  spec=importlib.util.find_spec(cv2)
  if spec is None:
      print(f"{cv2} is NOT installed.")
  else:
    print(f"{cv2} is installed.")
for lib in libraries:
    check_library(lib)
```
